```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
df = pd.read_csv('dummyft1.csv')  # Replace 'dummyft1.csv' with your actual CSV file name

# Group by 'tag' and sum the 'debit' for each tag
grouped_df = df.groupby('tag')['debit'].sum().reset_index()

# Extract columns from the grouped DataFrame
debit = grouped_df['debit'].tolist()
tags = grouped_df['tag'].tolist()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(debit, labels=tags, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

# Add a title
plt.title('Expenditure Distribution')

# Show the pie chart
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
df = pd.read_csv('dummyft1.csv')  # Replace 'dummyft1.csv' with your actual CSV file name

# Group by 'tag' and sum the 'debit' for each tag
grouped_df = df.groupby('tag')['debit'].sum().reset_index()

# Extract columns from the grouped DataFrame
debit = grouped_df['debit'].tolist()
tags = grouped_df['tag'].tolist()

# Create a bar graph
plt.figure(figsize=(10, 6))
plt.bar(tags, debit, color=plt.cm.Paired.colors)

# Add labels and title
plt.xlabel('Tags')
plt.ylabel('Total Debit')
plt.title('Expenditure Distribution')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the bar graph
plt.show()
```
